```python
# 열에 있는 값을 행으로 내리는 방법
# melt : 지정한 열에 있는 값을 행으로 내리는 방법
# pd.melt(df, id_vars = ...) id_vars에는 고정한 열을 입력한다.
# var_name : 위치를 변경한 열의 이름
# value_name : 위치를 변경할 열의 데이터를 저장한 열의 이름

import pandas as pd

df = pd.read_csv("pew.csv")

df2 = pd.melt(df, id_vars=["religion"], var_name="income", value_name = "count")
df2
```

```python
ebola = pd.read_csv("country_timeseries.csv")
ebola
ebola2 = pd.melt(ebola, id_vars=["Date","Day"])
```

```python
ebola2
```

```python
ebola_split = ebola2["variable"].str.split("_")
```

```python
state = ebola_split.str[0]
country = ebola_split.str[1]

ebola2["State"] = state
ebola2["Country"] = country

del ebola2["variable"]
ebola2

```
