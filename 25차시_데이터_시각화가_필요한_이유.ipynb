```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
data1 = anscombe[anscombe["dataset"] == "I"]
data2 = anscombe[anscombe["dataset"] == "II"]
data3 = anscombe[anscombe["dataset"] == "III"]
data4 = anscombe[anscombe["dataset"] == "IV"]
```

```python
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
ax1.scatter(data1["x"], data1["y"])
ax2.scatter(data2["x"], data2["y"])
ax3.scatter(data3["x"], data3["y"])
ax4.scatter(data4["x"], data4["y"])
```

```python
ax1.plot(data1["x"], data1["y"])
ax2.plot(data2["x"], data2["y"])
ax3.plot(data3["x"], data3["y"])
ax4.plot(data4["x"], data4["y"])
```

```python
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")
```

```python
fig.tight_layout()
```
