```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure() # 그림을 그리기 위한 도화지
```

```python
ax1 = fig.add_subplot(1,1,1) # 1행 1열 1번째 자리
ax1.hist(tips["total_bill"], bins=10)
```

```python
ax1.hist(tips["total_bill"], bins=30)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(tips["total_bill"], tips["tip"], c = tips["size"], s=tips["size"]*20, alpha=0.5)
ax1.set_xlabel("TOTAL BILL")
ax1.set_ylabel("TIP")
ax1.set_title("SCATTER")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax1.plot(tips["total_bill"])
male = tips[tips["sex"] == "Male"]
female = tips[tips["sex"]== "Female"]

ax1.boxplot([male["total_bill"], female["total_bill"]])
```
