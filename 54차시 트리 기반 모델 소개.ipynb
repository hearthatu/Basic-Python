# 트리 기반 모델 소개

- 결정 트리

> 스무고개 하듯이 예 또는 아니요 질문을 이어가며 학습하는 모델

> 첫 질문을 Root Node라고 하며, 뒤에 질문은 Node라고 표현

> 가장 마지막 노드를 Leaf Node라 부른다.


- 결정트리 학습과정

> 가장 잘 데이터를 구분할 수 있는 질문으로 기준을 나눈다.

> 과대 적합이 발생하기 쉽다.


- 불순도

> 불순도는 해당 범주 안에 서로 다른 데이터가 얼마나 섞여 있는지 뜻한다.


- 엔트로피

> 불순도를 수치적으로 나타낸 척도. 엔트로피가 높다는 것은 불순도가 높다는 뜻이다.

- 랜덤 포레스트

> 결정트리 학습에는 과대 적합되는 경향이 있다.

> 랜덤 포레스트는 다수의 결정트리로부터 분류 결과를 취해서 결론을 얻는 방법이다.

```
from sklearn.ensemble import RandomForestClassifier
```


````python
from sklearn.datasets import load_iris

iris = load_iris()
data, target = iris["data"], iris["target"]

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target)
````

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree

dt = DecisionTreeClassifier(max_depth=3)
dt.fit(train_input, train_target)
dt.score(test_input, test_target)
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plot_tree(dt, feature_names = iris["feature_names"],filled=True)
````

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(train_input, train_target)
rf.score(test_input, test_target)
````
