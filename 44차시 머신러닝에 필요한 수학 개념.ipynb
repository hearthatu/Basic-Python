- 이차함수

> y = a^2(a>0)

1. 위 그래프를 x축으로 p, y를 q만큼 평행이동 -> y = a(x-p)^2 + p
2. x축으로 대칭 -> y = -ax^2
3. 이차함수 x의 근은 y=0과 만나는 교점에 해당하므로 이차방정식 = 0의 해를 구하는 것과 같다.
