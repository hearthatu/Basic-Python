데이터 크롤링 :

인터넷은 큰 틀로 보면, HTML이라는 언어로 구성, Python으로 HTML을 다루어야 하며, 페이지 요청을 할 수 있어야한다.

라이브러리

requests : 페이지 요청 라이브러리

bs4의 BeautifulSoup : Pyhonㅇ로 HTML을 다루는 라이브러리

time : 안전하게 수집하는 방법

pandas : 수집한 데이터 엑셀화

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")
```

```python
url # 200 : 문제 없음... / 400번대 : 존재하지 않는 페이지 / 500번대 : 로그인 또는 잠금처리가 된 홈페이지
```

```python
url.text
```

```python
html = BeautifulSoup(url.text)
```

```python
current = int(html.find("span", class_ = "f_red").text.replace("회", ""))
current
```

```python
lotto = html.find("div", class_ = "lottonum").find_all("span")
del lotto[-2]
del lotto[-2]

box = []

for i in lotto:
    box.append(int(i.text))

box
```
